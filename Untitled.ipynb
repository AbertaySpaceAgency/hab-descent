## Equations of Motion

Before we can simulate how our payload moves on its way down, we need to write down the _equations of motion_ for our porblem: a series of mathematical expressions that define how an object moves based on its properties (shape, weight, size) and the forces that apply to it.

To do this, we use Newton's second law. It says that the acceleration (AKA the change of speed) of a solid at any one point is the sum of all forces that apply to it, divide by its mass:

$$a = \frac{\sum{F}}{m}$$

We also know that the acceleration is the rate of change of the speed. In maths language, we say that the acceleration is the _derivative_ of the speed. $\dot{v}$ means "the rate of change of v over time". Likewise, the speed of an object is the rate at which is position changes -- the speed is the derivative of the position (in our case, the vertical $v$ velocity is the derivative of the altitude $y$).

$$\ddot{y} = \dot{v} = a \frac{\sum{F}}{m}$$

### Forces in Atmospheric Flight

\begin{align*}
D &= \frac{1}{2}\rho v^{2} C_{D}S_{A} \\
\ddot{y} &= \frac{\rho \dot{y}^{2} C_{D}S_{A}}{2m} - g \\
\end{align*}

## Numerical Solution

So we have a problem: our position depends on our speed. Our speed depends on our acceleration, which depends on the speed and the position (since the density varies with altitude). Because all of our equations depend on each other, we cannot just solve them analytically.

Instead, we solve our system _numerically_.

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
#plt.style.use('bmh')
matplotlib.rcParams['figure.figsize'] = (8.0, 4.0)
#%matplotlib notebook
```

```python
## We define constants that we'll use in our simulation:

DENSITY_0 = 1.221
SCALE_HEIGHT = 8.570e3 # Scale height is about 8 km on Earth
EARTH_RADIUS = 6371e3
EARTH_MU = 3.986004418e14

PARACHUTE_DIAM = 0.8
BOX_WIDTH = 0.3
PARACHUTE_CD = 0.75
BOX_CD = 1.05
MASS = 1.0

# Simulation parameters

STEP_SIZE = 0.5
START_ALT = 40e3
PARACHUTE_DEPLOY_ALT = 40.0e3
PARACHUTE_REEF_TIME = 1.0

BOX_AREA = BOX_WIDTH*BOX_WIDTH
PARACHUTE_AREA = np.pi * (PARACHUTE_DIAM/2.0)**2
```

## Atmospheric Density & Altitude

```python
@np.vectorize
def density(altitude):
    """ Returns the atmospheric density for a given altitude.
    """
    return DENSITY_0 * np.exp(-altitude/SCALE_HEIGHT)

@np.vectorize
def gravity(altitude):
    return EARTH_MU / (EARTH_RADIUS+altitude)**2

# We can plot our density from 0 to 100km for example:
alts = np.arange(0, 100e3, 100)
rhos = density(alts)
gs = gravity(alts)

plt.figure(1)
plt.subplot(121)
plt.title('Density Gradient')
plt.xlabel('atmospheric density')
plt.ylabel('altitude')
plt.plot(rhos, alts)

plt.subplot(122)
plt.title('Gravity Gradient')
plt.xlabel('gravity $\\frac{m}{s^{-2}}$')
plt.ylabel('altitude')
plt.plot(gs, alts)
plt.tight_layout()
plt.show()
```

```python
# Drag equation
def drag(y, v, Cd, Sa):
    return 0.5 * density(y) * (v ** 2) * Cd * Sa

def system_drag(y, v, chute_mult):
    return drag(y, v, PARACHUTE_CD, chute_mult*PARACHUTE_AREA) + drag(y, v, BOX_CD, BOX_AREA)

def weight(y):
    return MASS * gravity(y)

def forces(y, v, chute_mult=1.0):
    return system_drag(y, v, chute_mult) - weight(y)
```

```python
def euler(state, chute_mult=1.0):
    accel = forces(state[0], state[1], chute_mult)/MASS
    velocity = state[1]+accel*STEP_SIZE
    position = state[0]+state[1]*STEP_SIZE
    return (position, velocity, accel, state[3]+STEP_SIZE)

def verlet(state, chute_mult=1.0):
#     next.position += state.velocity * dt + state.acceleration * halfdt * dt;
#     next.velocity += state.acceleration * halfdt;
#     next.acceleration = callback(next, mass);
#     next.velocity += next.acceleration * halfdt;
    pos = state[0] + (state[1] * STEP_SIZE) + (state[2] * STEP_SIZE * 0.5 * STEP_SIZE)
    vel = state[1] + state[2] * 0.5 * STEP_SIZE
    acc = forces(state[0], state[1], chute_mult)/MASS
    vel = vel + acc * 0.5 * STEP_SIZE
    return (pos, vel, acc, state[3]+STEP_SIZE)
```

```python
state = (START_ALT, 0, 0, 0)
alts = []
vels = []
accels = []
times = []
reefing = []
iterations = 0
max_vel = 0

chute_mult = 0.0
deploy = False

while state[0] > 0:
    iterations += 1
    state = verlet(state, chute_mult)
    
    if state[1] > max_vel:
        max_vel = state[1]
    
    if not deploy and state[0] <= PARACHUTE_DEPLOY_ALT:
        deploy = True
        
    if deploy and chute_mult < 1.0:
        chute_mult += STEP_SIZE / PARACHUTE_REEF_TIME
        chute_mult = chute_mult if chute_mult < 1.0 else 1.0
    
    
    reefing.append(chute_mult)
    alts.append(state[0])
    vels.append(state[1])
    accels.append(state[2])
    times.append(state[3])

plt.figure(2)
plt.title('Altitude')
plt.xlabel('time')
plt.ylabel('altitude')
plt.plot(times, alts)
plt.tight_layout()
plt.show()

plt.figure(3)
plt.title('Velocity (vs time)')
plt.xlabel('time')
plt.ylabel('velocity')
plt.plot(times, np.abs(vels))
plt.tight_layout()
plt.show()

plt.figure(4)
plt.title('Descent Profile')
plt.subplot(121)
plt.ylabel('altitude')
plt.xlabel('velocity')
plt.plot(np.abs(vels), alts)
plt.subplot(122)
plt.ylabel('altitude')
plt.xlabel('acceleration')
plt.plot(np.subtract(np.abs(accels), 9.81), alts)
plt.tight_layout()
plt.show()

print('Simulation ended after %d iterations' % iterations)
print('Impact Parameters:')
print('  Time elapsed:    %.0fs (%dmin)' % (state[3], int(state[3]/60)))
print('  Velocity:        %.2fm/s' % np.abs(state[1]))
print('---')
print('Max. Velocity:     %.2fm/s' % np.max(np.abs(vels)))
```
